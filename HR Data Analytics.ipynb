```python
fig,ax = plt.subplots(2,2, figsize=(10,10))
plt.suptitle("Understanding the distribution of various factors", fontsize=20)
sns.distplot(df['Age'], ax = ax[0,0])
ax[0][0].set_title('Distribution of Age')
sns.distplot(df['TotalWorkingYears'], ax = ax[0,1])
ax[0][1].set_title('Distribution of Total Working Years')
sns.distplot(df['YearsAtCompany'], ax = ax[1,0])
ax[1][0].set_title('Distribution of Years at company')
sns.distplot(df['YearsWithCurrManager'], ax = ax[1,1])
ax[1][1].set_title('Distribution of Years with Current Manager')
plt.show()
```

From the plots, we find that majority of the employees are in the age group of 30-40 years and have a total experience of around 8-10 years. They majority have spent around 2-6 years at the current organization and have spent 0-3 years in the current role they are in.

```python
#Monthly income from the rate of attrition according to gender
sns.barplot(x='Attrition', y='MonthlyIncome', hue= 'Gender',data=df)
plt.show()
```

```python
#Bar plot of Job Level with respect to Monthly income according to attrition
sns.barplot(x='Attrition', y='MonthlyIncome', hue= 'JobLevel',data=df)
```

```python
#Bar plot of Marital Status with respect to Monthly income according to attrition
sns.barplot(x='Attrition', y='MonthlyIncome', hue= 'MaritalStatus',data=df)
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x='Attrition', y='MonthlyIncome', hue= 'YearsAtCompany',data=df)
```

We can clearly observe the difference in monthly income with respect to years at the organisation. The difference in the monthly income is observed by the median value of attrition.

```python
f,ax = plt.subplots(figsize=(15,10))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, hue='JobLevel',palette='Set2')
plt.legend(loc='best')
plt.show()
```

We can clearly observe the difference in monthly income across different job levels. The difference in the monthly income is observed by the attrition.

```python
sns.jointplot(df.Age,df.MonthlyIncome, kind = "scatter")   
plt.show()
```

The scatter plot depicts the link between monthly income and age. A linear relationship can be discovered. Furthermore, the density map above depicts the age distribution, while the density plot on the right depicts the monthly income distribution.

```python
#Pairwise plots between continuous variables show the relationship between them
data = ['Attrition','Age','MonthlyIncome','DistanceFromHome']
sns.pairplot(df[data], kind="reg", diag_kind = "kde" , hue = 'Attrition' )
plt.show()
```

Observing the relationship between Age and Monthly Income, we can find that with age, monthly income has increased but the increase is not similar for attrition
